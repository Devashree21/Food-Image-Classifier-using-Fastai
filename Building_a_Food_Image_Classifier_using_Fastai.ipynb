<h3 style="font-family: Arial;background-color:#ffc2d1;color:black;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;letter-spacing: 2px;font-size: 20px"><strong><centre>Creating Labels to classify images </centre></strong></h3>

We are creating three labels with food names of our choice to classify the food images.

```python
labelA = 'cheesecake'
labelB = 'donuts'
labelC= 'panna_cotta'
```

```python
for img in get_image_files(foodPath):    
  
  if labelA in str(img):
    img.rename(f"{img.parent}/{labelA}-{img.name}")
  elif labelB in str(img):
    img.rename(f"{img.parent}/{labelB}-{img.name}")
  elif labelC in str(img):
    img.rename(f"{img.parent}/{labelC}-{img.name}")
  else: os.remove(img) 

len(get_image_files(foodPath))
```

```python
def GetLabel(fileName):
  return fileName.split('-')[0]
GetLabel("cheesecake-1092082.jpg") 
```

```python
dls = ImageDataLoaders.from_name_func(
    foodPath, get_image_files(foodPath), valid_pct=0.2, seed=42,
    label_func=GetLabel, item_tfms=Resize(224))

dls.train.show_batch()
```

<h3 style="font-family: Arial;background-color:#ffc2d1;color:black;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;letter-spacing: 2px;font-size: 20px"><strong><centre>Model Training </centre></strong></h3>

```python
learn = vision_learner(dls, resnet34, metrics=accuracy, pretrained=True)
learn.fine_tune(epochs=10)
```

Now, let us test a few sample images to check how our model performs.

```python
from google.colab import files
uploader = files.upload()

for img in uploader.items():
  uploadedImg = img[0]

img = PILImage.create(uploadedImg)
img.show()

label,_,probs = learn.predict(img)

print(f"This is a {label}.")
print(f"{labelA} {probs[0].item():.2f}")
print(f"{labelB} {probs[1].item():.2f}")
print(f"{labelC} {probs[2].item():.2f}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

for i in range(0,10):
  randomIndex = random.randint(0, len(get_image_files(foodPath))-1)
  img = mpimg.imread(get_image_files(foodPath)[randomIndex])
  label,_,probs = learn.predict(img)

  fig = plt.figure()
  ax = fig.add_subplot() 
  imgplot = plt.imshow(img) 
  ax.set_title(label) 

  plt.gca().axes.get_yaxis().set_visible(False)
  plt.gca().axes.get_xaxis().set_visible(False)
```
